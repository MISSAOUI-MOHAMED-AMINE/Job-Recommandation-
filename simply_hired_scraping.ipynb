```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException
import time

def job_details(job_url):
    # Set up Chrome options
    options = Options()
    options.headless = True  # Run in headless mode (no UI)
    driver = webdriver.Chrome(options=options)
    
    try:
        # Open the webpage
        driver.get(job_url)

        # Allow some time for the page to load
        time.sleep(3)

        # Scrape the relevant job details with try-except for each field
        try:
            poste = driver.find_element(By.CSS_SELECTOR, "div[class='chakra-stack css-1iblfv6']").text
        except NoSuchElementException:
            poste = "Non Spécifié"

        try:
            company = driver.find_element(By.CSS_SELECTOR, "span[data-testid='viewJobCompanyName']").text
        except NoSuchElementException:
            company = "Non Spécifié"

        try:
            location = driver.find_element(By.CSS_SELECTOR, "span[data-testid='viewJobCompanyLocation']").text
        except NoSuchElementException:
            location = "Non Spécifié"

        try:
            contract = driver.find_element(By.CSS_SELECTOR, "span[data-testid='viewJobBodyJobDetailsJobType']").text
        except NoSuchElementException:
            contract = "Non Spécifié"

        try:
            salary = driver.find_element(By.CSS_SELECTOR, "span[data-testid='viewJobBodyJobCompensation']").text
        except NoSuchElementException:
            salary = "Non Spécifié"

        try:
            pubdate = driver.find_element(By.CSS_SELECTOR, "span[data-testid='viewJobBodyJobPostingTimestamp']").text
        except NoSuchElementException:
            pubdate = "Non Spécifié"

        try:
            benefits = driver.find_element(By.CSS_SELECTOR, "div[data-testid='viewJobBodyJobBenefits']").text.split("\n")
        except NoSuchElementException:
            benefits = ["Non Spécifié"]  # Return a list with "Non Spécifié" for consistency

        try:
            qualif = driver.find_element(By.CSS_SELECTOR, "div[data-testid='viewJobQualificationsContainer']").text.split("\n")[1:]
        except NoSuchElementException:
            qualif = ["Non Spécifié"]

        # Collect the data into a list
        job_data = [company, poste, location, contract, salary, pubdate, benefits, qualif, job_url]

        return job_data

    finally:
        # Close the browser
        driver.quit()
```

```python

def get_links(job):
    # Initialize an empty list to store the job links
    links = []
    
    # Setup headless Chrome options
    options = Options()
    options.headless = True
    driver = webdriver.Chrome(options=options)
    
    try:
    
        job = job.replace(' ', '+')
        # Construct the URL based on the job query and page number
        link = f"https://www.simplyhired.com/search?q={job}&sr=0&s=d"
        
        # Open the URL
        driver.get(link)
        
        # Find all the job listing items
        items = driver.find_elements(By.CSS_SELECTOR, "li[class='css-0']")
        
        # Extract the links and append them to the list
        for item in items:
            links.append(item.find_element(By.TAG_NAME, 'a').get_attribute('href'))
    
    finally:
        # Close the driver after the scraping is done
        driver.quit()
    
    return links
```

```python
def save_jobs_to_dataframe():
    # Get all job links
    job_links = get_links("data science")
    # List to hold all job data
    job_data = []
    
    # Loop over each link and extract job info
    for link in job_links:
        info = job_details(link)
        job_data.append(info)
    
    # Create a DataFrame
    df = pd.DataFrame(job_data, columns=['company', 'poste', 'location', 'contract', 'salary', 'pubdate', 'benefits', 'qualif', 'job_url']
)
    
    # Save to CSV (optional)
    df.to_csv('simply_hired.csv', index=False)
    
    return df

# Example usage

```

```python
save_jobs_to_dataframe()
```
